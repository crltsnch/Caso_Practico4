Importamos librerías

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Cargamos datos.

```python
df = pd.read_csv('../data/data.csv', sep=';', encoding='latin1')
df
```

Ver datos vacíos

```python
df.isnull().sum()
```

```python
# Rellenar los valores nulos en las columnas 'Casado' y 'Graduado' con la moda de cada una
df['Casado'] = df['Casado'].fillna(df['Casado'].mode()[0])
df['Graduado'] = df['Graduado'].fillna(df['Graduado'].mode()[0])

# Verificar los resultados
print(df[['Casado', 'Graduado']].isnull().sum())


df.drop(['Campanna_anno'], axis=1, inplace=True)
df
```

```python
print(df.dtypes)
```

```python
# Reemplazar las comas por puntos y convertir la columna 'Digital_encuesta' a tipo numérico
df['Digital_encuesta'] = df['Digital_encuesta'].str.replace(',', '.').astype(float)

# Verificar la conversión
print(df['Digital_encuesta'].head())
print(df.dtypes)  # Para asegurarte de que la columna es ahora numérica
```

Pasamos de categóricas a numéricas.

```python
# Seleccionar las columnas categóricas
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
categorical_cols = [col for col in categorical_cols if col not in ['ID', 'Digital_encuesta']]

# Crear el LabelEncoder
le = LabelEncoder()

# Transformar las columnas categóricas y mostrar el mapeo
codificaciones = {}  # Diccionario para almacenar las codificaciones

for col in categorical_cols:
    le.fit(df[col])  # Ajustar el LabelEncoder a la columna
    df[col] = le.transform(df[col])  # Transformar la columna
    
    # Guardar el mapeo de valores originales y codificados
    codificaciones[col] = dict(zip(le.classes_, le.transform(le.classes_)))
    
 # Imprimir las codificaciones
for col, mapping in codificaciones.items():
    print(f"Codificación para '{col}': {mapping}")   

df
```

```python
print(df.dtypes)
```

Exportamos datos limpios.

```python
df.to_csv('../data/data_clean.csv', index=False)  # Sin índices
```
