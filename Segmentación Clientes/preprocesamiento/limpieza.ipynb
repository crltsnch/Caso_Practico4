Importamos librerías

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Cargamos datos.

```python
df = pd.read_csv('../data/data.csv', sep=';', encoding='latin1')
df
```

Ver datos vacíos

```python
df.isnull().sum()
```

```python
df.drop(['Casado', 'Graduado', 'Campanna_anno'], axis=1, inplace=True)
df
```

Pasamos de categóricas a numéricas.

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Crear el LabelEncoder
le = LabelEncoder()

# Transformar las columnas categóricas automáticamente
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

df
```

Exportamos datos limpios.

```python
df.to_csv('../data/data_clean.csv', index=False)  # Sin índices
```
